```python
import requests
from bs4 import BeautifulSoup
```

```python
base_url = ("http://www.pythonhow.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/t=0&s=")
l = []  #the dictionary data will be stored in this list
# we can use find_all("a",{"class",:"Page"}) to find last page number
```

```python
base_url
```

```python
for page in range(0,30,10):
    webpage = str(base_url+str(page)+".html")  # iterate through all the subpages
    print(webpage)
    r = requests.get(webpage)
    c = r.content
    soup = BeautifulSoup(c, "html.parser")
    all = soup.find_all("div", {"class":"propertyRow"})
    for item in all:
        d = {}  # all data will be stored in the dictionary
        d["Price"]=item.find_all("h4",{"class":"propPrice"})[0].text.replace("\n","").replace(" ","")
        d["Adress1"]=item.find_all("span",{"class":"propAddressCollapse"})[0].text
        d["Adress2"]=item.find_all("span",{"class":"propAddressCollapse"})[1].text
        try:
            d["Beds"]=item.find("span",{"class":"infoBed"}).find("b").text  # some data don't have bed numbers, so just pass it
        except:
            d["Beds"]="non"
    
        try:
            d["Area"]=item.find("span",{"class":"infoSqFt"}).find("b").text
        except:
            d["Area"]="non"
    
        try:
            d["FullBath"]=item.find("span",{"class":"infoValueFullBath"}).find("b").text
        except:
            d["FullBath"]="non"
    
        try:
            d["HalfBath"]=item.find("span",{"class":"infoValueHalfBath"}).find("b").text
        except:
            d["HalfBath"]="non"
        for column_group in item.find_all("div", {"class":"columnGroup"}):
            for feature_group, feature_name in zip(column_group.find_all("span",{"class":"featureGroup"}),
                                                   column_group.find_all("span",{"class":"featureName"})):
                #print(feature_group.text,feature_name.text)
                if "Lot Size" in feature_group.text:
                    d["LotSize"]=feature_name.text
                else:
                    d["LotSize"]="non"
        l.append(d)
```

```python
import pandas
df = pandas.DataFrame(l)  #create a dataframe of list "l"
# in dataframe, we can create matrix data like in Matlab
```

```python
df.to_csv("OutPut.csv")  # save all the results in a csv file
```
